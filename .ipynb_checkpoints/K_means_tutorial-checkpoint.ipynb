```python
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn import metrics
import numpy as np

import warnings
warnings.filterwarnings('ignore')

np.random.seed(33)
%matplotlib inline
```

```python
X, y = make_blobs(n_samples = 400, n_features = 2, centers = 7, cluster_std = 0.8, random_state = 33)
plt.figure(figsize = (10,10)) 
plt.scatter(X[:, 0], X[:, 1], c = y, s = 25)
plt.tight_layout();
```

```python
k_means = KMeans(n_clusters = 7, init = 'k-means++', algorithm = 'auto')
k_means.fit(X)
y_hat = k_means.predict(X)
```

```python
plt.figure(figsize = (10,10))
plt.scatter(X[:,0], X[:,1], c = y_hat, s = 25)
cl_centers = k_means.cluster_centers_
plt.scatter(cl_centers[:,0], cl_centers[:,1], c = 'black', s = 40)
plt.tight_layout()
```

```python
labels = k_means.labels_
metrics.silhouette_score(X, labels, 'euclidean')
```

```python
metrics.calinski_harabaz_score(X, labels)
```

```python
k_means6 = KMeans(n_clusters = 6, init = 'k-means++', algorithm = 'auto')
k_means6.fit(X)
y_hat6 = k_means6.predict(X)
```
